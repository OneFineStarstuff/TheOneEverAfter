```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from torchvision import datasets, transforms

# Define the VAE class (as before)
class VAE(nn.Module):
    def __init__(self, input_dim, hidden_dim, latent_dim):
        super(VAE, self).__init__()
        # Encoder
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)  # Mean
        self.fc_logvar = nn.Linear(hidden_dim, latent_dim)  # Log variance

        # Decoder
        self.fc3 = nn.Linear(latent_dim, hidden_dim)
        self.fc4 = nn.Linear(hidden_dim, input_dim)

    def encode(self, x):
        h1 = torch.relu(self.fc1(x))
        return self.fc_mu(h1), self.fc_logvar(h1)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        h3 = torch.relu(self.fc3(z))
        return torch.sigmoid(self.fc4(h3))

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar

# Define the loss function
def loss_function(recon_x, x, mu, logvar):
    # Reconstruction loss (binary cross-entropy)
    recon_loss = nn.functional.binary_cross_entropy(recon_x, x, reduction='sum')

    # KL divergence
    kl_div = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())

    return recon_loss + kl_div

# Define EWC loss with additional debugging
def ewc_loss(new_model, old_model, lam=0.5):
    loss = 0
    for (name, new_param), (old_name, old_param) in zip(new_model.named_parameters(), old_model.named_parameters()):
        # Print parameter shapes for debugging
        print(f"Comparing {name}: new_param shape {new_param.shape} vs old_param shape {old_param.shape}")
        if new_param.shape == old_param.shape:
            loss += (lam / 2) * (new_param - old_param).pow(2).sum()
        else:
            print(f"Skipping {name} due to shape mismatch: {new_param.shape} vs {old_param.shape}")
    return loss

# Hyperparameters
input_dim = 784  # Example for MNIST
hidden_dim = 400
latent_dim = 20
num_epochs = 10
batch_size = 128
learning_rate = 1e-3

# Load and normalize MNIST dataset
transform = transforms.Compose([transforms.ToTensor(), transforms.Lambda(lambda x: x.view(-1))])
train_dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Initialize the VAE and optimizer
vae = VAE(input_dim, hidden_dim, latent_dim)
old_vae = VAE(input_dim, hidden_dim, latent_dim)
old_vae.load_state_dict(vae.state_dict())
optimizer = optim.Adam(vae.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    total_loss = 0
    for x, _ in train_loader:
        recon_x, mu, logvar = vae(x)
        vae_loss = loss_function(recon_x, x, mu, logvar)

        # Incorporate EWC loss
        ewc_loss_value = ewc_loss(vae, old_vae, lam=0.5)
        total_loss_value = vae_loss + ewc_loss_value

        optimizer.zero_grad()
        total_loss_value.backward()
        optimizer.step()

        total_loss += total_loss_value.item()

    avg_loss = total_loss / len(train_loader.dataset)
    print(f'Epoch {epoch+1}, Loss: {avg_loss:.4f}')

print("Training complete.")
```
